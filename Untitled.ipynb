# now we use GRADIENT DESCENT to find the best fitting line

SSE- Sum of Squared Error
There are severals ways of measure the error, we will be using SSE 

```python
numInstances = 41
def SSE(m,b):
    
    totalError = 0.0
    
    for i in range(numInstances):
        
        x = X[i]
        y = Y[i]
        
        #The real rating
        currentTarget = y
        
        #Predicted rating with our current fitting line
        #y = mx + b
        currentOutput = m*x + b
        
        #Compute squared error
        currentSquaredError = (currentTarget - currentOutput)**2

        
        #Add it to the total error
        totalError += currentSquaredError

    sse = totalError/numInstances
    
    return totalError
```

```python
m = 2
b = 3

sse = SSE(m,b)

print('For the fitting line: y = %sx + %s\nSSE: %.2f' %(m,b,sse))
```

# Minimizing our error - Gradient Descent

```python
def gradient_descent_step(m,b):
    
    N = numInstances
    m_grad = 0
    b_grad = 0
    
    for i in range(N):

        #Get current pair (x,y)
        x = X[i]
        y = Y[i]

        #Partial derivative respect 'm'
        dm = -((2/N) * x * (y - (m*x + b)))
        
        #Partial derivative respect 'b'
        db = - ((2/N) * (y - (m*x + b)))
    
        #Update gradient
        m_grad = m_grad + dm
        b_grad = b_grad + db
    
    #Set the new 'better' updated 'm' and 'b'
    m_updated = m - 0.0001*m_grad
    b_updated = b - 0.0001*b_grad
    
    return m_updated,b_updated
```

```python
m = 2
b = 3

sse = SSE(m,b)

print('Starting line: y = %.2fx + %.2f - Error: %.2f' %(m,b,sse))
```

```python
for i in range(2):
    m,b = gradient_descent_step(m,b)
    sse = SSE(m,b)
    print('At step %d - Line: y = %.2fx + %.2f - Error: %.2f' %(i+1,m,b,sse))

print('\nBest  line: y = %.2fx + %.2f - Error: %.2f' %(m,b,sse))
```

```python
def gradient_descent_n_steps(m_starting,b_starting,steps): #For doing it many times in an easy way ;)
    print('Starting line: y = %.6fx + %.6f - Error: %.6f\n' %(m_starting,b_starting,SSE(m_starting,b_starting)))
    m = m_starting
    b = b_starting
    display_freq = steps//10
    for i in range(steps):
        m,b = gradient_descent_step(m,b)
        if(i % display_freq == 0):
            sse = SSE(m,b)
            print('At step %d - Line: y = %.6fx + %.6f - Error: %.6f' %(i+1,m,b,sse))
    
    
    print('\nBest  line: y = %.6fx + %.6f - Error: %.6f' %(m,b,sse))
    return m,b
```
